## 为什么从数据结构开始？

AI 编程的本质是**处理数据**。不管是训练神经网络还是构建 RAG 系统，你都在做同一件事：

1. 把数据**装进**某种容器（列表、字典）
2. 对数据做**变换**（筛选、计算、格式化）
3. 把结果**传给**下一个环节

Python 的列表、字典和函数就是完成这三步的基本工具。掌握它们，后面学 NumPy、PyTorch 时才不会卡壳。

**学习目标**:
- 掌握列表（List）的创建、筛选、变换
- 掌握字典（Dict）的键值存储和遍历
- 理解列表推导式（AI 代码中出现频率极高）
- 学会定义函数，理解参数和返回值

## 1. 列表 (List) — 最常用的数据容器

### 为什么要学列表？

在 AI 中，数据几乎总是以"一组数字"的形式存在：
- 一张图片 = 一组像素值 `[128, 64, 255, ...]`
- 一个句子 = 一组词的编号 `[1024, 512, 8, ...]`
- 一批训练数据 = 一组样本 `[样本1, 样本2, ...]`

列表就是 Python 中最基本的"一组数据"容器。

### 下面这段代码在做什么？

我们创建一个成绩列表，然后用 Python 内置函数快速计算统计信息。
重点观察：`max()`、`min()`、`sum()`、`len()` 这四个函数，后面会反复用到。

````python
# 创建一个成绩列表（7 个学生的分数）
scores = [85, 92, 78, 95, 88, 76, 90]

print(f"原始成绩: {scores}")
print(f"最高分: {max(scores)}")
print(f"最低分: {min(scores)}")
print(f"平均分: {sum(scores) / len(scores):.1f}")
````

### 你学到了什么？

- `[85, 92, 78, ...]` 用方括号创建列表
- `max()` / `min()` / `sum()` / `len()` 是处理数字列表的四大金刚
- `f"...{表达式}"` 是 Python 的格式化字符串，`:.1f` 表示保留 1 位小数

> 💡 **和 AI 的关系**：训练神经网络时，你会不断计算一批数据的平均损失（loss），用的就是 `sum(losses) / len(losses)` 这个模式。

### 列表推导式 — AI 代码中的高频语法

### 为什么要学这个？

列表推导式是 Python 的"一行代码批量处理"语法。在 AI 代码中出现频率极高，比如：
- 筛选出损失值大于阈值的样本
- 对每个数据点做归一化
- 把一批文本转成小写

如果你看不懂列表推导式，后面的代码会处处卡壳。

### 下面这段代码在做什么？

两个操作：
1. **筛选**：从成绩中挑出 >= 90 分的（条件过滤）
2. **变换**：给每个成绩加 5 分（批量运算）

````python
# 筛选：只保留 >= 90 分的成绩
high_scores = [s for s in scores if s >= 90]
print(f"90分以上: {high_scores}")

# 变换：每个成绩加 5 分
boosted = [s + 5 for s in scores]
print(f"加5分后: {boosted}")
````

### 你学到了什么？

列表推导式的两种模式：

```python
# 模式1：变换（对每个元素做操作）
[表达式 for 变量 in 列表]

# 模式2：筛选（只保留满足条件的）
[表达式 for 变量 in 列表 if 条件]
```

等价的普通写法（更长但更好理解）：
```python
# [s + 5 for s in scores] 等价于：
boosted = []
for s in scores:
    boosted.append(s + 5)
```

> 💡 **和 AI 的关系**：后面学 NumPy 时你会发现，NumPy 数组可以直接写 `scores + 5` 实现同样的效果，比列表推导式更快。但理解推导式是基础。

## 2. 字典 (Dict) — 给数据贴标签

### 为什么要学字典？

列表的问题是：`scores[3]` 是第 4 个成绩，但你不知道这是谁的成绩。

字典用"键-值"对解决这个问题：`{"小明": 95}` 一目了然。

在 AI 项目中，字典无处不在：
- **模型配置**：`{"learning_rate": 0.001, "batch_size": 32}`
- **训练日志**：`{"epoch": 5, "loss": 0.23, "accuracy": 0.91}`
- **API 返回值**：`{"model": "gpt-4", "content": "..."}`

### 下面这段代码在做什么？

模拟一个 AI 模型的配置参数。这和你后面用 PyTorch 训练模型时的参数设置方式几乎一样。

````python
# 模拟一个 AI 模型的配置参数
model_config = {
    "model_name": "my_first_model",
    "learning_rate": 0.001,     # 学习率：每次更新权重的步长
    "batch_size": 32,           # 批大小：每次训练用多少个样本
    "epochs": 10,               # 轮数：整个数据集训练几遍
    "hidden_layers": [128, 64, 32],  # 隐藏层：每层有多少个神经元
}

# 用 key 取值
print(f"模型名称: {model_config['model_name']}")
print(f"学习率: {model_config['learning_rate']}")

# 遍历所有配置
print("\n完整配置:")
for key, value in model_config.items():
    print(f"  {key}: {value}")
````

### 你学到了什么？

- `{"key": value}` 用花括号创建字典
- `dict["key"]` 通过键取值
- `dict.items()` 遍历所有键值对
- 字典的值可以是任何类型：字符串、数字、甚至列表

> 💡 **和 AI 的关系**：上面的 `model_config` 不是随便编的。`learning_rate`、`batch_size`、`epochs` 是训练神经网络时最重要的三个超参数，后面的练习你会真正用到它们。

### 字典推导式 — 快速构造字典

### 下面这段代码在做什么？

把之前的成绩列表变成一个"学生→成绩"的字典。`enumerate()` 函数同时给出索引和值。

````python
# 把成绩列表变成 {学生编号: 成绩} 的字典
student_scores = {f"学生{i+1}": score for i, score in enumerate(scores)}
print(f"学生成绩字典: {student_scores}")
````

### 你学到了什么？

- `enumerate(列表)` 返回 `(索引, 值)` 的配对，索引从 0 开始
- 字典推导式语法：`{key表达式: value表达式 for 变量 in 列表}`
- `f"学生{i+1}"` 用格式化字符串动态生成键名

> 💡 **和 AI 的关系**：后面做 RAG 时，你会把文档块存成 `{"chunk_0": "文本内容...", "chunk_1": "..."}` 这样的结构，用的就是这个模式。

## 3. 函数 — 把重复的操作打包

### 为什么要学函数？

假设你有 10 组不同的数据，每组都要算最大值、最小值、平均值。
不用函数 → 同样的代码复制粘贴 10 遍。
用函数 → 写一次，调用 10 次。

在 AI 项目中，函数是组织代码的基本单位：
- `train()` — 训练模型
- `evaluate()` — 评估模型
- `predict()` — 用模型预测

### 下面这段代码在做什么？

定义一个 `calculate_stats()` 函数，输入一组数字，输出一个包含统计信息的字典。
注意看函数签名中的**类型标注**（`: list[float]` 和 `-> dict`），这是 Python 的好习惯。

````python
def calculate_stats(data: list[float]) -> dict:
    """计算一组数据的统计信息

    Args:
        data: 数字列表

    Returns:
        包含统计信息的字典
    """
    return {
        "count": len(data),
        "mean": sum(data) / len(data),
        "max": max(data),
        "min": min(data),
        "range": max(data) - min(data),
    }


stats = calculate_stats(scores)
print(f"成绩统计: {stats}")
````

### 你学到了什么？

- `def 函数名(参数) -> 返回类型:` 定义函数
- `"""..."""`  三引号字符串是文档注释（docstring），说明函数的用途
- 函数可以返回字典，把多个结果打包在一起
- 定义好函数后，用 `函数名(参数)` 调用

> 💡 **和 AI 的关系**：后面训练神经网络时，你会写类似的函数来计算训练指标（损失、准确率等），结构和这个一模一样。

### 归一化函数 — AI 中最重要的数据预处理

### 为什么要学这个？

这是本练习中**最重要**的一个函数，因为"归一化"是 AI 的核心操作之一。

**用一个真实场景理解：** 假设你要让 AI 预测一个人的房贷是否会批准，输入两个特征：

```
          年龄    年收入（元）
张三       25      50000
李四       35      800000
王五       28      120000
```

年龄的范围是 25~35，年收入的范围是 50000~800000。收入的数字比年龄大了几万倍。

神经网络内部做的是加权求和：`结果 = w1 × 年龄 + w2 × 收入`

即使 w1 和 w2 一样大，收入的影响也会完全碾压年龄——因为 `800000 × w` 远大于 `35 × w`。模型会觉得"年龄根本不重要"，但实际上年龄很重要。

**归一化就是把所有特征缩放到同一个量级：**

```
归一化到 [0, 1]：

          年龄    年收入
张三       0.00    0.00      ← 两列各自的最小值变成 0
李四       1.00    1.00      ← 两列各自的最大值变成 1
王五       0.30    0.09      ← 中间值按比例缩放
```

现在年龄和收入都在 0~1 之间，谁也压不过谁，模型可以公平地学习两个特征各自的重要性。

### 下面这段代码在做什么？

定义一个 `normalize()` 函数，对列表里的**每一个数字**都用同一个公式缩放。
注意 `min_val=0, max_val=1` 是**默认参数**，调用时可以不传。

````python
def normalize(data: list[float], min_val: float = 0, max_val: float = 1) -> list[float]:
    """将数据归一化到指定范围

    公式: new_value = min_val + (x - data_min) / (data_max - data_min) * (max_val - min_val)

    Args:
        data: 原始数据
        min_val: 目标最小值（默认 0）
        max_val: 目标最大值（默认 1）

    Returns:
        归一化后的数据
    """
    data_min = min(data)
    data_max = max(data)
    return [
        min_val + (x - data_min) / (data_max - data_min) * (max_val - min_val)
        for x in data
    ]


# 归一化到 [0, 1]
normalized = normalize(scores)
print(f"原始成绩: {scores}")
print(f"归一化到 [0,1]: {[f'{x:.2f}' for x in normalized]}")

# 归一化到 [-1, 1]
normalized_neg = normalize(scores, -1, 1)
print(f"归一化到 [-1,1]: {[f'{x:.2f}' for x in normalized_neg]}")
````

### 你学到了什么？

- **默认参数**：`min_val=0` 表示调用时可以不传这个参数，默认用 0
- 观察输出：原来最低分 76 变成了 0.00，最高分 95 变成了 1.00

**归一化公式拆解**（以年收入 `[50000, 800000, 120000]` 归一化到 [0,1] 为例）：

```
data_min = 50000
data_max = 800000
范围 = 800000 - 50000 = 750000

公式：(x - data_min) / (data_max - data_min)

50000  → (50000  - 50000) / 750000 = 0 / 750000     = 0.00  ← 最小值变成 0
800000 → (800000 - 50000) / 750000 = 750000 / 750000 = 1.00  ← 最大值变成 1
120000 → (120000 - 50000) / 750000 = 70000 / 750000  = 0.09  ← 中间值按比例
```

代码里的列表推导式 `[(x - data_min) / (data_max - data_min) for x in data]` 就是对列表里的每一个数字都套一遍这个公式。

> 💡 **和 AI 的关系**：
> - 训练神经网络前，几乎总要对输入数据做归一化
> - 图片像素值 0-255 → 归一化到 0-1 或 -1 到 1
> - 后面学 PyTorch 时你会看到 `transforms.Normalize((0.5,), (0.5,))` 做的就是这件事
> - 不做归一化 → 数值大的特征碾压数值小的 → 模型学偏 → 效果差
